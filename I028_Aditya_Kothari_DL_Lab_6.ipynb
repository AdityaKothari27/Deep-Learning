```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from keras import models,layers,losses,metrics,regularizers
```

# Step - 1

```python
df = pd.read_csv('/content/drive/MyDrive/IRIS_dataset.csv')
```

```python
df
```

# Step - 2

```python
X = df[['sepal_length', 'sepal_width', 'petal_length', 'petal_width']].values

Y = df['species'].values
```

```python
from sklearn.preprocessing import LabelEncoder
```

```python
label_encoder = LabelEncoder()
Y = label_encoder.fit_transform(Y)
```

```python
Y
```

```python
from sklearn.preprocessing import StandardScaler
```

```python
scaler = StandardScaler()
X = scaler.fit_transform(X)
```

```python
X
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_val, Y_train, Y_val = train_test_split(X, Y, test_size=0.3, random_state=42)
```

# Step - 3

```python
import tensorflow as tf
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense
```

```python
model = Sequential()

model.add(Dense(64, input_dim=4, activation='relu'))
model.add(Dense(32, activation='relu'))
model.add(Dense(3, activation='softmax'))
```

# Step - 4

```python
model.compile(loss='sparse_categorical_crossentropy', optimizer='adam', metrics=['accuracy'])
```

```python
model.fit(X_train, Y_train, epochs=20, batch_size=16, validation_data=(X_val, Y_val))
```

```python
from tensorflow.keras.optimizers import SGD
```

```python
sgd_optimizer = SGD(learning_rate=0.1)
model.compile(loss='sparse_categorical_crossentropy', optimizer=sgd_optimizer, metrics=['accuracy'])

history = model.fit(X_train, Y_train, epochs=20, batch_size=16, validation_data=(X_val, Y_val))
```

```python
plt.figure(figsize=(10, 6))
plt.plot(history.history['accuracy'], label=f'Training Accuracy')
plt.plot(history.history['val_accuracy'], label=f'Validation Accuracy')
plt.title(f'Training and Validation Accuracy (lr={0.1})')
plt.xlabel('Epochs')
plt.ylabel('Accuracy')
plt.ylim(0.95, 1)
plt.legend()
plt.grid(True)
plt.show()

max_val_accuracy = max(history.history['val_accuracy'])
print(f"For learning rate {0.1}, max validation accuracy: {max_val_accuracy:.4f}")
```

# Step - 6

```python
# Define a list of momentum values to try
momentum_values = [0, 0.5, 0.9, 0.99]

# Store results
best_momentum = None
best_val_accuracy = 0

for momentum in momentum_values:
    model = Sequential()

    model.add(Dense(64, input_dim=4, activation='relu'))
    model.add(Dense(32, activation='relu'))
    model.add(Dense(3, activation='sigmoid'))

    sgd_optimizer = tf.keras.optimizers.SGD(learning_rate=0.01, momentum=momentum)
    model.compile(loss='sparse_categorical_crossentropy', optimizer=sgd_optimizer, metrics=['accuracy'])

    history = model.fit(X_train, Y_train, epochs=20, batch_size=16, validation_data=(X_val, Y_val))

    final_val_accuracy = history.history['val_accuracy'][-1]

    print(f"\nFor momentum {momentum},\n Final validation accuracy : {final_val_accuracy:.4f},\n")

    if final_val_accuracy > best_val_accuracy:
        best_val_accuracy = final_val_accuracy
        best_momentum = momentum

print(f"\nThe best momentum value is {best_momentum} with validation accuracy :  {best_val_accuracy:.4f}")
```

Best momentum = 0.9

# Step - 7

```python
momentum = 0.9
decay_values = [1E-1, 1E-2, 1E-3, 1E-4]

best_decay = None
best_val_accuracy = 0

for decay in decay_values:

    model = Sequential()

    model.add(Dense(64, input_dim=4, activation='relu'))
    model.add(Dense(32, activation='relu'))
    model.add(Dense(3, activation='sigmoid'))
    sgd_optimizer = tf.keras.optimizers.legacy.SGD(
        learning_rate=0.01, momentum=momentum, decay=decay
    )
    model.compile(loss='sparse_categorical_crossentropy', optimizer=sgd_optimizer, metrics=['accuracy'])

    history = model.fit(X_train, Y_train, epochs=20, batch_size=16, validation_data=(X_val, Y_val))

    final_val_accuracy = history.history['val_accuracy'][-1]
    print(f"\nFor decay {decay:.1e},\nFinal validation accuracy: {final_val_accuracy:.4f}\n")

    if final_val_accuracy > best_val_accuracy:
        best_val_accuracy = final_val_accuracy
        best_decay = decay

print(f"\nThe best decay value is {best_decay:.1e}\n")
print(f"The best validation accuracy is {best_val_accuracy:.4f}")
```

Best decay value = 1E-4

# Step - 8

```python
lr = 0.01
momentum = 0.9
decay = 1E-3
optimizers = ['Adagrad', 'Adam', 'RMSprop']
best_optimizer = None
best_val_accuracy = 0

for optimizer_name in optimizers:
    model = Sequential()

    model.add(Dense(64, input_dim=4, activation='relu'))
    model.add(Dense(32, activation='relu'))
    model.add(Dense(3, activation='softmax'))

    if optimizer_name == 'Adagrad':
        optimizer = tf.keras.optimizers.legacy.Adagrad(learning_rate=lr, decay=decay)
    elif optimizer_name == 'Adam':
        optimizer = tf.keras.optimizers.Adam(learning_rate=lr, beta_1=0.9, beta_2=0.999, epsilon=1e-07, amsgrad=False)
    elif optimizer_name == 'RMSprop':
        optimizer = tf.keras.optimizers.RMSprop(learning_rate=lr, rho=0.9, epsilon=1e-07)

    model.compile(loss='sparse_categorical_crossentropy', optimizer=optimizer, metrics=['accuracy'])

    history = model.fit(X_train, Y_train, epochs=20, batch_size=16, validation_data=(X_val, Y_val))
    final_val_accuracy = history.history['val_accuracy'][-1]

    print(f"\nFor optimizer {optimizer_name},\nFinal validation accuracy: {final_val_accuracy:.4f}\n")

    if final_val_accuracy > best_val_accuracy:
        best_val_accuracy = final_val_accuracy
        best_optimizer = optimizer_name

print(f"\nThe best optimizer is {best_optimizer}")
print(f"\nThe best validation accuracy is {best_val_accuracy:.4f}")
```

The best optimizer is Adam
