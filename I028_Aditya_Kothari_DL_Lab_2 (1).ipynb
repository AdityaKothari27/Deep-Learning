```python
import pandas as pd
import numpy as np
from sklearn.preprocessing import StandardScaler
```

```python
df1 = pd.get_dummies(df['species'])
```

```python
df1.head()
```

```python
df.drop('species',axis=1,inplace=True)
```

```python
df2 = pd.concat([df,df1],axis=1)
df2.head()
```

```python
columns = ['sepal_length','sepal_width','petal_length','petal_width']
```

```python
scaler = StandardScaler()
```

```python
df2.iloc[:,[0,1,2,3]] = scaler.fit_transform(df2.iloc[:,[0,1,2,3]])
```

```python
df2.head()
```

```python
np.random.seed(42)
w_i_h1=np.random.rand(4,1)
w_i_h2=np.random.rand(4,1)
w_i_h3=np.random.rand(4,1)
w_h_o1=np.random.rand(2,1)
w_h_o2=np.random.rand(2,1)
w_h_o3=np.random.rand(2,1)
bias1=np.random.rand(1)
bias2=np.random.rand(1)
bias3=np.random.rand(1)
```

```python
def sigmoid(x):
  return 1/(1+np.exp(-x))

input = df2.iloc[:,0:4]
```

```python
Z2_1 = np.dot(input,w_i_h1) + bias1
Z2_2 = np.dot(input,w_i_h2) + bias2
Z2_3 = np.dot(input,w_i_h3) + bias3
```

```python
h2_1 = sigmoid(Z2_1)
h2_2 = sigmoid(Z2_2)
h2_3 = sigmoid(Z2_3)
```

```python
h2=np.append(h2_1,h2_2,axis=1)
```

```python
Z2_1 = np.dot(h2,w_h_o1)
Z2_2 = np.dot(h2,w_h_o2)
Z2_3 = np.dot(h2,w_h_o3)
```

```python
o1=sigmoid(Z2_1)
o2=sigmoid(Z2_2)
o3=sigmoid(Z2_3)
```

```python
f1=np.array(df2.iloc[:,4])
f1=f1.reshape((-1,1))
e1=f1-o1
```

```python
f2=np.array(df2.iloc[:,4])
f2=f2.reshape((-1,1))
e2=f2-o2
```

```python
f3=np.array(df2.iloc[:,4])
f3=f3.reshape((-1,1))
e3=f3-o3
```

```python
error=np.multiply(e1,e2)+np.multiply(e2,e3)+np.multiply(e3,e1)
```

```python
error_total=np.sum(error)
error_total
```
